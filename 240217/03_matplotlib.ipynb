```python
# 라이브러리 로드 
import matplotlib.pyplot as plt
# from matplotlib import pyplot as plt
```

```python
## 라인 그래프 표시 
## plot({1차원 데이터})
## 1차원 데이터를 인자 값에 하나 대입 : y축 데이터 입력 , x축 데이터는 0부터 1씩 증가하는 데이터
## 1차원 데이터를 인자 값에 2개 대입 : 첫번째 인자 x축 데이터, 두번째 인자 y축 데이터 

plt.plot([5,3,7,1])
plt.show()
```

```python
plt.plot([1,2,3,4], [5,3,7,1])
plt.plot([1,3,5,7], [1,7,3,5])
plt.show()
```

```python
## 라인그래프의 선의 색을 변경 
plt.plot([3,5,2,1], '#FF0066')
plt.show()
```

```python
# 라인그래프의 선의 종류 
plt.plot([1,2,3,4], '--r')
plt.show()
```

```python
## x, y축의 라벨설정 
## 그래프의 제목 추가 
plt.plot([1,2,3,4], [6,3,2,8])
plt.xlabel('A')
plt.ylabel('B')
plt.title('Graph')
plt.show()
```

```python
# 막대 그래프 
# bar({x축 데이터}, {y축 데이터})
x = [1,2,3,4]
y = [10, 40, 30, 70]

plt.bar(x, y)
plt.show()
```

```python
plt.barh(x, y)
plt.show()
```

```python
plt.bar(x, y, color=['red', 'blue'], width=0.3)
plt.show()
```
